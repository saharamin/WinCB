## NAVER Movie Scraper (08.13.2018)

- 네이버 movie 사이트에서 영화정보를 가져옴 (영화 제목, 각 영화 사이트의 링크, 제목, 점수 등)
- 1. 몇 페이지의 정보를 가져올지 정하고
- 2. 나열된 각각의 영화 사이트 주소를 가져와 list로 저장
- 3. 이 list에서 저장된 각각의 영화 사이트로 가서 영화 정보를 dictionary로 저장

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url_list = []
npages=1 # 처리할 페이지 수
for page_num in range(npages):
    html = urlopen ('https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&page='+str(page_num+1))
    html_parsed = BeautifulSoup(html, 'html.parser')
  
    movie_list = html_parsed.findAll('div', {'class':'tit5'})
    for movie in movie_list:
        for link in movie.findAll('a', href=True):
            print("Found the URL:", link['href'])
            url_list.append(link['href'])

print('Number of urls:',len(url_list))
```

```python
movie_info_collection = []

count = 0 
print("total link:", len(url_list))

for movie_url in url_list:
    
    if count % 5 ==0:
        print(str(count),"/", len(url_list))    
        
    # 영화 link 마다 페이지 정보 parsing
    html = urlopen('https://movie.naver.com'+movie_url)
    movie = BeautifulSoup(html, 'html.parser')
 
    # 영화의 utl, 순위, id 저장하기
    movie_info = {}
    movie_info['url'] = 'https://movie.naver.com'+movie_url
    movie_info['ranking']=count+1
    movie_info['movie_id'] = movie_url.split('code=')[-1]
    
    # 영화의 title 가져오기
    title = movie.find('h3',{'class':'h_movie'})
    movie_info['title'] = title.find('a').text
    #movie_info['title'] = title.find('a').string    
    
    # 관람객 평점
    ntz_score = movie.find('a',{'class':'ntz_score'})
    movie_info['audience_rating'] = ntz_score.find('span',{'class':'st_on'}).text.split(" ")[-1].split("점")[0]
 
    # 기자, 평론가 평점
    try:
        a = movie.find('div', {'class': 'spc_score_area'})
        score = a.find('div', {'class': 'star_score'})
        em = score.findAll('em')
        movie_info['expert_rating']=""
        for i in range(len(em)):        
            movie_info['expert_rating']=movie_info['expert_rating']+ em[i].string        
    except:
        movie_info['expert_rating']="없음"
            
    try:
        a = movie.find('a',{'id':'pointNetizenPersentBasic'})
        b = a.findAll('em')
        movie_info['netizen_rating']=""
        for i in range(len(b)):        
            movie_info['netizen_rating']=movie_info['netizen_rating']+ b[i].string        
    except:
        movie_info['netizen_rating']="없음"
           
    count +=1
    movie_info_collection.append(movie_info)

print("Done")
```
